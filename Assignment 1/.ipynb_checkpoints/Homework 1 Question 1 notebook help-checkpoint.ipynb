# ISLR lab 3.6.2

```python
boston = pd.read_csv('Data/boston.csv')
boston.info()
```

```python
lm = sm.OLS.from_formula('medv ~ lstat', boston)
result = lm.fit()
result.summary()
```

Let's get the confidence intervals for the predicted coefficients

```python
result.conf_int()
```

Now we predict on some new data

```python
new = pd.DataFrame([[1, 5], [1, 10], [1, 15]], columns=['Intercept', 'lstat'])
preds = result.get_prediction(new)
preds.summary_frame(0.05)
```

Now we plot the regression line. Note that seaborn calls its own regression estimator here, so we are not using the sm.OLS model.

```python
sns.regplot(boston.lstat, boston.medv, scatter_kws={'color':'r', 's':8})
```

Now we plot the <b>residuals against the fitted values</b>

```python
fitted_values = pd.Series(result.fittedvalues, name="Fitted Values")
residuals = pd.Series(result.resid, name="Residuals")
sns.regplot(fitted_values, residuals, fit_reg=False)
```

Let's also try to plot the studentized (aka. pearson studentized) residuals

```python
studentized_residuals = pd.Series(result.resid_pearson, name="Studentized Residuals")
sns.regplot(fitted_values, studentized_residuals,  fit_reg=False)
```

Finally let's plot high leverage points

```python
from statsmodels.stats.outliers_influence import OLSInfluence
leverage = pd.Series(OLSInfluence(result).influence, name = "Leverage")
sns.regplot(leverage, studentized_residuals,  fit_reg=False)
```

# ISLR 3.6.3

Similar to the single-predictor regression, we can also do regression with multiple predictors using sm.OLS using the `from_formula` function to get syntax similar to that from R.

```python
result = sm.OLS.from_formula('medv ~ lstat + age', boston).fit()
result.summary()
```

Now let's try to <b>regress on all 13 predictors.</b> In order to avoid writing out all the column names using `from_formula` we will use a different syntax for constructing our OLS estimate. Note that we drop the `medv` column before passing `boston` as a predictor (what do you think would happen if we did not do this? Try it!).

```python
result = sm.OLS(boston.medv, boston.drop('medv', axis=1)).fit()
result.summary()
```

Let's try dropping the `age` column as well.

```python
result = sm.OLS(boston.medv, boston.drop(['medv', 'age'], axis=1)).fit()
result.summary()
```

# 3.6.4 - 3.6.5: Interactions Terms and Non-linear Transformations

Just as in R, we can easily include interaction terms.

```python
sm.OLS.from_formula('medv ~ lstat*age', boston).fit().summary()
```

We can also esaily do non-linear transformation:

```python
lm.fit2 = sm.OLS.from_formula('medv ~ lstat + np.square(lstat)', boston).fit()
lm.fit2.summary()
```

Let's use `anova` to quantify how much the squared predictor helps:

```python
lm.fit = sm.OLS.from_formula('medv ~ lstat', boston).fit()
sm.stats.anova_lm(lm.fit, lm.fit2)
```

Let's plot the <b>fittes values vs the residuals</b> again

```python
sns.regplot(lm.fit2.fittedvalues, lm.fit2.resid, fit_reg=False)
# No pattern found in this plot, 
# compared to the non-linear pattern found in the model with only 'lstat' as the predictor.
```

Now let's try a <b>fifth-order polynomial regression</b>:

```python
from sklearn.preprocessing import PolynomialFeatures
polynomial_features= PolynomialFeatures(degree=5)
x = boston.lstat.values.reshape(-1,1)
xp = polynomial_features.fit_transform(x)
sm.OLS(boston.medv, xp).fit().summary()
```
