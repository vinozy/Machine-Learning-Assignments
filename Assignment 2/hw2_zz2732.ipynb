### Part (c) Confusion Matrix from Logistic Regression

```python
features = ['Lag1', 'Lag2', 'Lag3', 'Lag4', 'Lag5', 'Volume']
response = 'Direction'
X = df[features]
y = df[response]
logreg = LogisticRegression(penalty='none') 
logreg.fit(X, y)
y_pred = logreg.predict(X)
df_confusion = pd.crosstab(y, y_pred)
df_confusion = pd.crosstab(y, y_pred, rownames=['Actual'], colnames=['Predicted'], margins=True)
display(df_confusion)
```

0 stands for 'Down' and 1 stands for 'Up'.

```python
# percentage of 'Up' in 'Direction' of all data points.
df['Direction'].mean()
```

```python
# Overall accuracy:
oa_c = (df_confusion[0][0]+df_confusion[1][1])/df_confusion['All']['All']
oa_c
```

```python
# False positive rate:
fp_c = df_confusion[1][0]/df_confusion['All'][0]
fp_c
```

A very high False Positive rate, not a good sign. 80% of stocks that actually went down were predicted going up.

```python
# True positive rate(sensitivity):
tp_c = df_confusion[1][1]/df_confusion['All'][1]
tp_c
```

### Part 1

Given the unconstrained maximization problem: $$\underset{\alpha}{max}\space \frac{\alpha^TB\alpha}{\alpha^TW\alpha}$$

Using the $scale$ $invariance$ of the Rayleigh quotient, rewrite into a constrained maximization problem: $$\underset{\alpha}{max}\space \alpha^TB\alpha \\ s.t. \space \alpha^TW\alpha = 1$$

Solve the optimization problem above, using Lagrange Multipliers.<br>
First, define the Lagrangian form, with $\lambda$ being the Lagrange Multiplier:
$$
L(\alpha,\lambda)=\alpha^TB\alpha + \lambda(\alpha^TW\alpha-1)
$$

Now, take partial derivatives with respect to $\alpha$ and $\lambda$, and set them equal to $0$:
$$
\frac{\partial L(\alpha,\lambda)}{\partial \alpha} = 2B\alpha + 2\lambda W \alpha = 0\\
\space\\
\frac{\partial L(\alpha,\lambda)}{\partial \lambda} = \alpha^TW\alpha - 1 = 0\\
$$

Solve the 1st equation ang get:
$$
-B\alpha = \lambda W \alpha \\
\\
-W^{-1}B\alpha=\lambda\alpha
$$
<br>
We get an eigenvalue problem, in eigen decomposition form.

The optimal solution $a^*$ will be the eigenvector corresponding to the matrix $-W^{-1}B$ with the largest eigenvalue $\lambda$.

From in-class lecture slides, we derived that the $l_th$ discriminant variable is 
$$
Z_l=v_l^TD^{-\frac{1}{2}}U^Tx
$$
<br>
Therefore, the $1_{st}$ discriminant variable is equal to $v_1^TD^{-\frac{1}{2}}U^Tx$

Since $W=\sum$ <br>
By eigen-decomposition, $W=(W^{1/2})^TW^{1/2}$<br>
Compute $B^* = (W^{-\frac{1}{2}})^TBW^{-\frac{1}{2}}$ <br>
The discriminant coordinates are $a_l=W^{-{1/2}}v_l^*$

Therefore, essentially by finding the lienar combination $Z=a^TX$ such that between-class variance is maximized relative to the within-class variance. We are finding the optimal $a^*$, such that $a^*x$ is the 1st discriminant variable $Z_1$ above.

### Part 1

For a binary logistic regression, we have derived the following in class:
$$
P(Y=1|X) = \frac{exp(\beta_{10}+\beta_{k}^Tx)}{1+\sum_{\iota=1}^{2}exp(\beta_{\iota0}+\beta_{\iota}^Tx)}
$$
<br>
$$
P(Y=0|X) = \frac{1}{1+\sum_{\iota=1}^{2}exp(\beta_{\iota0}+\beta_{\iota}^Tx)}
$$

If we predict $Class$ $1$ when $P(Y=1|X) >P(Y=0|X) $, it is equivalent to: $\frac{P(Y=1|X)}{P(Y=1|X)}>1$.

Try to prove it is a linear classifier via $monotone$ $transformation$ by taking $log$ on both sides to get the log-ratio between 2 classes:
$$
log(\frac{P(Y=1|X)}{P(Y=0|X)})>0
$$

Expand the term on the left side and we get:
$$
log(exp(\beta_{10}+\beta_{k}^Tx) - log(1+\sum_{\iota=1}^{2}exp(\beta_{\iota0}+\beta_{\iota}^Tx)) -log(1) + log(1+\sum_{\iota=1}^{2}exp(\beta_{\iota0}+\beta_{\iota}^Tx)) >0
$$

The 3rd term $log(1) = 0$. The 2nd and the last term cancelled out, so we get:

$$
log(exp(\beta_{10}+\beta_{k}^Tx)>0
$$

Equivalently, we get the classifier below:<br>
Predicting $Class$ $1$ when

$$
\beta_{10}+\beta_{k}^Tx > 0
$$
<br>
Therefore, the decision boundary is $\beta_{10}+\beta_{k}^Tx$ = 0 shows that logistic regression yields a linear classifier. 

### Part 2

Define the prior probabilities over classes $Y$, for k = 0,1:
$$
P(Y=1)=\pi, P(Y=0)=1-\pi
$$

Derive the parametric forms of $P(Y|X)$ under the Gaussian class-conditional probabilities:<br>
<br>
$$
P(Y=1|X)=\frac{P(Y=1)P(X|Y=1)}{P(Y=1)P(X|Y=1)+P(Y=0)P(X|Y=0)}
$$
<br>
$$
P(Y=1|X)=\frac{1}{1+exp(ln\frac{P(Y=0)P(X|Y=0)}{P(Y=1)P(X|Y=1)})}
$$
<br>
$$
P(Y=1|X)=\frac{1}{1+exp(ln\frac{P(Y=0)}{P(Y=1)}+\sum_jln(\frac{P(X_j|Y=0)}{P(X_j|Y=1)}))}
$$
<br>
$$
P(Y=1|X)=\frac{1}{1+exp(ln\frac{1-\pi}{\pi}+\sum_jln(\frac{P(X_j|Y=0)}{P(X_j|Y=1)}))}
$$

Substitute Gaussian PDF's into the conditional probabilities:<br>
<br>
$$
P(Y=1|X)=\frac{1}{1+exp(ln\frac{1-\pi}{\pi}+\sum_j(\frac{\mu_{j0}-\mu_{j1}}{\sigma_j^2}X_j+\frac{\mu_{j0}^2-\mu_{j1}^2}{2\sigma_j^2}))}
$$

Let $w_j = \frac{\mu_{j0}-\mu_{j1}}{\sigma_j^2}$ for $j=1,2,...,n$
and $w_0 = ln\frac{1-\pi}{\pi}+\sum_j\frac{\mu_{j0}^2-\mu_{j1}^2}{2\sigma_j^2}$

We then have:

<br>
$$
P(Y=1|X)=\frac{1}{1+exp(w_0+\sum_{j=1}^{n}w_jX_j)}
$$
<br>
$$
P(Y=0|X)=\frac{exp(w_0+\sum_{j=1}^{n}w_jX_j)}{1+exp(w_0+\sum_{j=1}^{n}w_jX_j)}
$$

Similarly, set the log-odds of these 2 probabilities to $0$ and solve to get the decision boundary:
<br>
$$
log(\frac{P(Y=1|X)}{P(Y=0|X)})=0
$$
<br>
$$
log(1)-log(1+exp(w_0+\sum_{i=j}^{n}w_jX_j))-log(exp(w_0+\sum_{j=1}^{n}w_jX_j))+log(1+exp(w_0+\sum_{j=1}^{n}w_jX_j))=0
$$
<br>
$$
log(exp(w_0+\sum_{j=1}^{n}w_jX_j))=0
$$
<br>
$$
w_0+\sum_{j=1}^{n}w_jX_j=0
$$
<br>
Therefore, the decision boundary is $w_0+\sum_{j=1}^{n}w_jX_j=0$, a linear classifier in the same parametric form as Logistic Regression above.

### Part 3

By comparing the 2 classifiers, the parameters in Logistic Regression can be expressed in terms of the parameters $w_j$ in the Naive Bayes.

The optimal parameters $\beta$ in Logistic Regression is computed iteratively in the following method:
<br>
$$
\beta^{new} = argmin_\beta(Z-X_\beta)^TW(Z-X_\beta)
$$
<br>
Where $W$ is the diagonal matrix with $P(X_j;\beta)(1-P(X_j;\beta))$ for $j=1,2,...,n$ on the diagonal.

Yes, if the assumptions of Gaussian class-conditional probabilities hold, the Naive Bayes and Logistic Regression classifiers will converge towards a same classifier as the training sample size grows to infinity.
